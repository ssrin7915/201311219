```python
%%writefile web_data_8_paging.py
import scrapy
from urlparse import urljoin

class MyItem(scrapy.Item):
    text = scrapy.Field()
    author = scrapy.Field()

class MySpider(scrapy.Spider):
    name = "quotes"
    start_urls = [ 'http://quotes.toscrape.com/page/2/']
    #start page
    
    def parse(self, response):
        #print len(response.xpath('//div[@class="quote"]'))
    
        for quote in response.xpath('//div[@class="quote"]'):
            item = MyItem()
            item['text'] = quote.xpath('span[@class="text"]/text()').extract_first() 
            item['author'] = quote.xpath('span/small/text()').extract_first()
            #print item['text'],"by",item['author']
            print item['author']
            yield item
        
        n_page = response.xpath('//li[@class="next"]/a/@href').extract_first()
        if n_page: 
            n_page= urljoin('http://quotes.toscrape.com',n_page) #http://quotes.toscrape.com/page/?/
            print ""
            print "--> NEXT PAGE "
            print ""
            yield scrapy.Request(n_page, callback=self.parse) 
```
