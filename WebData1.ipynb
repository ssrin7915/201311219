## 웹데이터-1: python.org 페이지를 크롤링해서http url 출력하기##

```python
import urllib
```

```python
response = urllib.urlopen('http://python.org/')
html = response.read()
```

# regex #

```python
import re
```

```python
p=re.compile('href="(http://.*?)"')  #href=http:// 로 시작하는 것은 무엇이든
```

```python
res=p.findall(html) #html안에서 찾기
```

```python
for i in res:
    print i
```

# xpath #

```python
import lxml
import lxml.etree
```

```python
tree=lxml.etree.HTML(html)
```

```python
nodes=tree.xpath('//*[@href]') #href 속성을 가진 모든 태그를 반환
```

```python
for node in nodes:
    print node.tag, node.attrib 
```

# CSS selector #

```python
import lxml.html
```

```python
from lxml.cssselect import CSSSelector
```

```python
import requests
```

```python
r =requests.get('http://python.org/')
```

```python
html = lxml.html.fromstring(r.text)
```

```python
sel = CSSSelector('a[href]') #모든 a태그의 href 속성
```

```python
results = sel(html)
```

```python
for i in results:
    print lxml.html.tostring(i)
```

```python
for x in results:
    print x.get('href'), x.text #href 값과 그 a태그 텍스트 받아오기
```
