### 신한카드 -  전체받아오기 ###

```python
import urllib
from urlparse import urljoin
import codecs
response=urllib.urlopen('https://www.shinhancard.com/conts/person/card_info/major/benefit/propose/propose.jsp')
_html=response.read()

from lxml import etree
tree=etree.HTML(_html)
shinhan='https://www.shinhancard.com'

pages = tree.xpath('//*[@id="pbContent"]/div[@class="tabWrap1"]/ul/li/ul/li/a/@href')
cn=0
dn=0

f=codecs.open('SinhancardBenefit.txt','w','utf-8')
f.write(u'---------신한카드----------')
f.write('\n')
for page in pages:
    url = urljoin(shinhan,page)
    url_tree = etree.HTML(urllib.urlopen(url).read())
    nodes = url_tree.xpath('//*[@id="pbContent"]/div[@class="cardList"]')
    for node in nodes:
        cn += 1
        title=node.xpath('div[@class="cardName"]/strong/text()')[0]
        detail=node.xpath('div[@class="btn"]/a[@class="btnCardWhite"]/@href')[0]
        detail_url = urljoin(shinhan,detail)
        detail_tree = etree.HTML(urllib.urlopen(detail_url).read())
        #print title
        #print detail_url
        for x in detail_tree.xpath('//*[@id="pbTabDepth1-1"]/div'):
            for benefit in x.xpath('p/text()'):
                dn += 1
                f.write(benefit) #print "*",benefit
                f.write('\n')
f.close()
```

```python
print cn,dn
```

### 우리카드 ###

```python
import urllib
from urlparse import urljoin
response=urllib.urlopen('https://sccd.wooribank.com/ccd/Dream?withyou=cd')
_html=response.read()

from lxml import etree
tree=etree.HTML(_html)
woori = "https://sccd.wooribank.com"
import codecs
f=codecs.open('WooricardBenefit.txt','w','utf-8')

cardList = tree.xpath('//*[@id="lnb"]/ul/li[2]/div/div/ul/li[1]/dl/dd/a/@href')
cn=0
dn=0
for card in cardList:
    url = urljoin(woori,card)
    url_tree = etree.HTML(urllib.urlopen(url).read())
    nodes = url_tree.xpath('//*[@id="content"]/div[@class="card-display-box"]/div/div/div/div/div/div[1]/div')
    for node in nodes:
        cn += 1
        title=node.xpath('h3/text()')[0]
        #print "#",title
        details=node.xpath('div[@class="list-cd"]/ul/li/text()')
        for detail in details:
            dn += 1
            f.write(detail)
            print detail
            f.write('\n')
print "---------"

tree2=etree.HTML(urllib.urlopen('https://sccd.wooribank.com/ccd/Dream?withyou=CDCIF0021&CTGR_CD=C200012').read())
cardList2 = tree2.xpath('//*[@id="frm"]/div[1]/ul/li/a/@href')
for card in cardList2:
    url = urljoin(woori,card)
    url_tree = etree.HTML(urllib.urlopen(url).read())
    nodes = url_tree.xpath('//*[@id="frm"]/div[3]/div')
    for node in nodes:
        cn+=1
        title=node.xpath('div[1]/div[2]/dl/dt/a/text()')[0]
        #print "#",title
        details=node.xpath('div[2]/ul/li/p[2]/text()')
        for detail in details:
            dn += 1
            f.write(detail)
            print detail
            f.write('\n')
f.close()
```

### 삼성카드 ###

```python
import urllib
from urlparse import urljoin
from lxml import etree
import codecs

f=codecs.open('SamsungcardBenefit.txt','w','utf-8')
cn=0
dn=0
samsung = "https://www.samsungcard.com"
tree=etree.HTML(urllib.urlopen('https://www.samsungcard.com/personal/card/digital/UHPPCC0388M0.jsp').read())
cardList = tree.xpath('//*[@id="contents"]/div/div[3]/ul/li')
for card in cardList:
    title=card.xpath('div/a[1]/span[1]/text()')[0]
    cn += 1
    #print "#",title 
    detail=card.xpath('div/a[1]/@href')[0]
    detail_url = urljoin(samsung,detail)
    detail_tree = etree.HTML(urllib.urlopen(detail_url).read())
    for detail in detail_tree.xpath('//*[@id="benefit00"]/ul/li'):
        for benefit in detail.xpath('a/dl/dd/div/ul/li/span/text()'):
            dn += 1
            f.write(benefit)
            f.write('\n')
            print benefit
print "---------"

tree2=etree.HTML(urllib.urlopen('https://www.samsungcard.com/personal/card/number/UHPPCA0202M0.jsp').read())
cardList2 =tree2.xpath('//*[@id="contents"]/div/div/div/ul/li')
for card in cardList2:
    for detail in card.xpath('div/ul/li'):
        title=detail.xpath('div[2]/p/a/text()')[0]
        cn += 1
        #print "#",title
        url=detail.xpath('div[2]/div/a[1]/@href')[0]
        detail_url = urljoin(samsung,url)
        detail_tree = etree.HTML(urllib.urlopen(detail_url).read())
        for x in detail_tree.xpath('//*[@id="benefit00"]/ul/li'):
            for benefit in x.xpath('a/dl/dd/div/ul/li/span/text()'):
                dn += 1
                f.write(benefit)
                f.write('\n')
                print benefit
print "---------"


tree3=etree.HTML(urllib.urlopen('https://www.samsungcard.com/personal/card/check-card/UHPPCA0206M0.jsp').read())
cardList3 =tree3.xpath('//*[@id="card_append"]/li')
for card in cardList3:
    title = card.xpath('dl/dt/a/text()')[0]
    cn += 1
    #print "#",title
    detail = card.xpath('dl/dd[2]/a/@href')[0]
    detail_url="https://www.samsungcard.com/personal/card/cardfinder/UHPPCA0102M0.jsp?code="+detail[25:32]
    detail_tree = etree.HTML(urllib.urlopen(detail_url).read())
    for x in detail_tree.xpath('//*[@id="benefit00"]/ul/li'):
        for benefit in x.xpath('a/dl/dd/div/ul/li/span/text()'):
            dn += 1
            f.write(benefit)
            f.write('\n')
            print benefit
f.close()
```

```python
import oauth2
import os
import urllib
import json

def getKey(keyPath):
    d=dict()
    f=open(keyPath,'r')
    lines=f.readlines()
    for line in lines:
        row=line.split('=')
        row0=row[0]
        d[row0] =row[1].strip()
    return d

keyPath=os.path.join(os.getcwd(),'src','key.properties')

key=getKey(keyPath)

consumer = oauth2.Consumer(key=key['api_key'], secret=key['api_secret'])
token=oauth2.Token(key=key['access_token'], secret=key['access_secret'])
client = oauth2.Client(consumer,token)
```
